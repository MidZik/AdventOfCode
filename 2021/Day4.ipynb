# Part 1

```python
# run to solve part 1
class Board:
    def __init__(self, board_data):
        self.data = [[int(item) for item in row.split()] for row in board_data]
        self.marked = [[False for _ in row] for row in self.data]
    
    def mark_number(self, number):
        for row, col in product(range(5), repeat=2):
            if self.data[row][col] == number:
                self.marked[row][col] = True
    
    def is_winner(self):
        for row in range(5):
            if all(self.marked[row]):
                return True
        for col in range(5):
            if all(self.marked[row][col] for row in range(5)):
                return True
        return False
    
    def sum_unmarked(self):
        return sum(self.data[row][col] for row, col in product(range(5), repeat=2) if not self.marked[row][col])
        

def part_1(input):
    lines = input.splitlines()
    number_draws = lines[0].split(',')
    boards = []
    for board_data in chunked(lines[1:], 6):
        boards.append(Board(board_data[1:]))
    
    for number in number_draws:
        number = int(number)
        for board in boards:
            board.mark_number(number)
            if board.is_winner():
                return board.sum_unmarked() * number

part_1_answer = part_1(aoc.get_input(2021, 4))
print(part_1_answer)
```

# Part 2

```python
# run to solve part 2

def part_2(input):
    lines = input.splitlines()
    number_draws = lines[0].split(',')
    boards = []
    for board_data in chunked(lines[1:], 6):
        boards.append(Board(board_data[1:]))
    
    last_winner_score = None
    last_winner_number = None
    
    for number in number_draws:
        number = int(number)
        for board in boards:
            if board.is_winner():
                continue
            board.mark_number(number)
            if board.is_winner():
                last_winner_score = board.sum_unmarked()
                last_winner_number = number
    
    return last_winner_score * last_winner_number
part_2_answer = part_2(aoc.get_input(2021, 4))
print(part_2_answer)
```

```python
# run to submit part 2

aoc.submit_answer(2021, 4, 2, part_2_answer)
```
