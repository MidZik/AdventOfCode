# Part 1

```python
# run to solve part 1

def part_1(aoc_input):
    lines = aoc_input.splitlines()
    lines_iter = iter(lines)
    dots = set()
    folds = []
    for line in lines_iter:
        if line == "":
            break
        x, y = map(int, re.findall(r"(\d+),(\d+)", line)[0])
        dots.add((x, y))
        print(x, y)
    for line in lines_iter:
        axis, val = re.findall(r"fold along (\w)=(\d+)", line)[0]
        print(axis, val)
        folds.append((axis, int(val)))
    
    for axis, val in folds:
        for x, y in list(dots):
            match axis:
                case 'x':
                    if x >= val:
                        dots.discard((x, y))
                    if x > val:
                        diff = x - val
                        dots.add((val - diff, y))
                case 'y':
                    if y >= val:
                        dots.discard((x, y))
                    if y > val:
                        diff = y - val
                        dots.add((x, val - diff))
        print(dots)
        return len(dots)
    
    return len(dots)

part_1_test_answer = part_1(test_input)
print("TEST RESULT: ", part_1_test_answer)
part_1_answer = part_1(aoc.get_input(2021, 13))
print("ACTUAL RESULT: ", part_1_answer)
```

# Part 2

```python
# run to solve part 2

def part_2(aoc_input):
    lines = aoc_input.splitlines()
    lines_iter = iter(lines)
    dots = set()
    folds = []
    for line in lines_iter:
        if line == "":
            break
        x, y = map(int, re.findall(r"(\d+),(\d+)", line)[0])
        dots.add((x, y))
    for line in lines_iter:
        axis, val = re.findall(r"fold along (\w)=(\d+)", line)[0]
        folds.append((axis, int(val)))
    
    for axis, val in folds:
        for x, y in list(dots):
            match axis:
                case 'x':
                    if x >= val:
                        dots.discard((x, y))
                    if x > val:
                        diff = x - val
                        dots.add((val - diff, y))
                case 'y':
                    if y >= val:
                        dots.discard((x, y))
                    if y > val:
                        diff = y - val
                        dots.add((x, val - diff))

    max_x = max(x for x, y in dots)
    max_y = max(y for x, y in dots)
    for y in range(max_y + 1):
        for x in range(max_x + 1):
            if (x, y) in dots:
                print('#', end='')
            else:
                print(' ', end='')
        print()

part_2_test_answer = part_2(test_input)
print("TEST RESULT: ", part_2_test_answer)
part_2_answer = part_2(aoc.get_input(2021, 13))
print("ACTUAL RESULT: ", part_2_answer)
```
